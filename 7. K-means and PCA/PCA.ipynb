```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat('ex7data1.mat')
X = data['X']
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1])
plt.show()
```

```python
def PCA(X):
    X = (X - X.mean()) / X.std()
    cov = (X.T @ X) / X.shape[0]
    u, s, v = np.linalg.svd(cov)
    return u, s, v
```

```python
U, S, V = PCA(X)
U, S, V
```

```python
def projectData(X, U, k): # k为所降成的维度
    Ureduced = U[:,:k]
    return X@Ureduced

def recoverData(Z, U, k):
    Ureduced = U[:,:k]
    return Z@Ureduced.T
```

```python
Z = projectData(X, U, 1)
```

```python
Z
```

```python
Xrecovered = recoverData(Z, U, 1)
Xrecovered
```

```python
data = loadmat('ex7faces.mat')
faces = data['X']
```

```python
def plotFaces(faces):
    size = int(np.sqrt(faces.shape[1]))
    index = np.random.randint(0, len(faces), 25)
    temp = faces[index]
    fig, ax = plt.subplots(nrows = 5, ncols = 5, sharex = True, sharey = True, figsize = (8, 8))
    for i in range(5):
        for j in range(5):
            ax[i][j].imshow(temp[i * 5 + j].reshape((size, size)), cmap = 'gray')
    plt.xticks([])
    plt.yticks([])
    plt.show()
plotFaces(faces)
```

```python
U, S, V = PCA(faces)
Z = projectData(faces, U, 100)
facesReduced = recoverData(Z, U, 100)
plotFaces(facesReduced)
```
