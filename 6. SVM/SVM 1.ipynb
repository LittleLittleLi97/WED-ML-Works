```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn import svm
```

```python
data = loadmat('ex6data1.mat')
X = data['X']
y = data['y']
```

```python
fig, ax = plt.subplots()
for i in range(len(y)):
    if y[i] == 1:
        ax.scatter(X[i][0], X[i][1], marker = '+', color = 'r')
    else:
        ax.scatter(X[i][0], X[i][1], color = 'b')
plt.show()
```

```python
svc = svm.LinearSVC(C = 1, loss = 'hinge', max_iter = 1000) # C就是cost里面的C，hinge是那个cost的函数，max_iter是迭代次数
# multi_class训练的类型 'ovr' one versus rest 一对多（多种的分类）
# tol 误差小于这么多后就停止
# penalty costReg那部分
svc.fit(X, y.flatten())
```

```python
svc.score(X, y.flatten())
```

```python
fig, ax = plt.subplots()
ax.scatter(X[ : , 0], X[ : , 1], c = svc.decision_function(X), cmap = 'seismic')
plt.show()
# decistion_function(X) 每个点的置信度
```
